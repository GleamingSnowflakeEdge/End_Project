```python
%%writefile app.py
import streamlit as st
import os
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains.question_answering import load_qa_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from PyPDF2 import PdfReader

# Set environment variable
os.environ['HUGGINGFACEHUB_API_TOKEN'] = "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"

# Set web page title and icon
st.set_page_config(
    page_title="Chat with PDF",
    page_icon=":robot:"
)

# Set web page title and markdown
st.title('Chat with PDF (Powered by Open-Source LLM)')
st.markdown(
    """
    This is the demonstration of a chatbot with PDF using an Open-source LLM, Chroma, and Streamlit.
    """
)

# Upload PDF file
pdf = st.file_uploader("Upload your PDF", type='pdf')

if pdf is not None:
    pdf_reader = PdfReader(pdf)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text()

    # Chunk the text
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len
    )
    chunks = text_splitter.split_text(text)

    # Create embeddings
    embeddings = HuggingFaceEmbeddings(model_name="google/flan-t5-base")
    persist_directory = 'db'
    vectordb = Chroma.from_texts(chunks, embedding=embeddings, persist_directory=persist_directory)
    vectordb.persist()

    # Set up LLM pipeline
    model_id = 'google/flan-t5-large'
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_id)
    pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer, max_length=512)

    # Get input text
    input_text = st.text_input("Ask a question about your PDF:")

    if input_text:
      # Retrieve relevant documents
      docs = vectordb.similarity_search(input_text, k=3)

      # Set up and run QA chain
      local_llm = HuggingFacePipeline(pipeline=pipe)
      qa_chain = load_qa_chain(local_llm, chain_type="stuff")

      try:
          response = qa_chain({"input_documents": docs, "question": input_text})
          if isinstance(response, dict) and 'output_text' in response:
              st.write(response['output_text'])
          elif isinstance(response, str):
               st.write(response)
          else:
              st.write("Unexpected response format. Here's the raw output:")
              st.write(response)
      except Exception as e:
          st.error(f"An error occurred: {str(e)}")
```
